```python
import json
from bqplot import pyplot as plt
from bqplot import *
```

```python
with open("../Data/Visualization_Ready_Datasets/topojson_worldmap/cow1946_topo.json", "r") as read_file:
    cow1946_map_data = json.load(read_file)
```

```python
sampleJan = cow1946_map_data['1946-01-01']
```

```python
sampleJan['objects']['subunits']['geometries']
```

```python
map_fig = plt.figure(title="CoW 1946", min_width=1300, min_height=800)

map_tt = Tooltip(fields=['country', 'war_count'], labels=['country', 'war_count'])
map_res = plt.geo(map_data=sampleJan, stroke_color='black',
                 colors={'default_color': 'Gray'},
                 scales={'projection': Mercator(),
                        'color': ColorScale(scheme='magma')},
                 tooltip=map_tt)
```
