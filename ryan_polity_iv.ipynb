The "Polity IV Annual Time-Series, 1800-2018" dataset is about the authority characteristics of political regimes around the world. The dataset consists of 17562 rows and 36 columns. In our exploratory analysis, we aim to provide a synthetic visualization of the general tendency of each political regime in the dataset to be autocratic or democratic. There are two columns in the dataset that are useful for our exploration:

1. `country`: The names of the political regimes coded in the dataset.
2. `polity2`: The standardized score times-series of the political regimes coded that ranks the degree of institutionalized autocracy/democracy from -10 (most autocratic) to 10 (most democratic).

Dataset Citation: "Polity IV Annual Time-Series, 1800-2018." Political Regime Characteristics and Transitions, 1800-2018, Polity IV Project, Center for Systemic Peace. http://www.systemicpeace.org/inscr/p4v2018.xls.

```python
# Import libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

We first import the libraries we will use in our exploratory analysis. For our purpose, we will use `pandas` to read in the dataset and `matplotlib` to visualize our analysis.

```python
# Import the dataset.
df1 = pd.read_csv('./Data/PolityIV/Raw/p4v2018.csv')
```

Here we use the function `read_csv` to import the dataset. Note that the dataset has been transformed into a `csv` file before being imported.

```python
# View the dataset.
df1
```

We then looked at the data frame to make sure the dataset had been imported correctly.

```python
# Select columns.
df2 = df1[['country', 'polity2']]
```

Since our dataset has been imported correctly, we can select the columns we need for our analysis, which are the `country` column and the `polity2` column.

```python
# View the new dataset.
df2
```

Again, we want to make sure our selection is successful by looking at the data frame.

```python
# Sort country names alphabetically.
df3 = df2.sort_values(by='country')
```

Although the above cell looks nice, we soon notice that the names of the political regimes are not listed in an alphabetical order. To help the audience better understand our analysis, we need to sort the data frame alphabetically by the `country` column.

```python
# View the new dataset.
df3
```

Now we look at the data frame again to double check everything is done correctly.

```python
# Plot the data.
plt.figure(figsize=(120, 10))
plt.scatter(df3['country'], df3['polity2'], c=df3['polity2'], cmap='RdYlGn', edgecolor='gray', s=100)
plt.colorbar()
plt.xticks(rotation='vertical', fontsize=20)
plt.title('Polity Scores by Country', fontsize=30)
plt.xlabel('Country', fontsize=25)
plt.ylabel('Polity Score', fontsize=25)
plt.show()
```

Then we can move on to visualizing the data we have selected. As we want to construct a visualization that provides an overview of the different polities that each political regime in the dataset has been through, we use a scatter plot mapped onto a diverging color scheme to show the polity change of each political regimes in the dataset. The visualization shows the different polity scores that each political regime receives and clearly indicates the position that each political regime is on the spectrum identified therein. For best results, please open the visualization in a new tab.

```python
# Group polity scores by country and calculate their mean by country.
df4 = df3.groupby('country')['polity2'].mean().reset_index()
```

We then want to take a step further by calculating the mean of the polity scores of each political regime so that we can provide a simpler but equally informative visualization.

```python
# View the new dataset.
df4
```

Once we have done with the calculation, we also need to look at the resulting data frame to be sure if everything is correct.

```python
# Plot the data.
plt.figure(figsize=(120, 10))
plt.bar(df4['country'], df4['polity2'])
plt.xticks(rotation='vertical', fontsize=20)
plt.title('Polity Score Mean by Country', fontsize=30)
plt.xlabel('Country', fontsize=25)
plt.ylabel('Polity Score Mean', fontsize=25)
plt.show()
```

Finally, we can provide another visualization that we want for our data. Now that each political regime has only one score to visualize, we decide to use a bar chart to visualize our data for better visual effects. By rendering our visualization this way, we are able to easily differentiate between political regimes that receive positive polity score means and are more democratic and those that receive negative polity score means and are more autocratic generally. Again, please open the visualization in a new tab for best results.
